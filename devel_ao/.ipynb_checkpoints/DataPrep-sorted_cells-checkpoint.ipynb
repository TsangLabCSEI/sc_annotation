```r
library(Seurat)
library(dplyr)
library(tibble)
library(stringr)
library(SeuratDisk)
```

```r
# read in raw matrix 
adt_raw = read.table("./data/sorted_cells.adt.raw.txt", sep = '\t', header = T, row.names=1, stringsAsFactors = F)
gex_raw = read.table("./data/sorted_cells.gex.raw.txt", sep = '\t', header = T, row.names=1, stringsAsFactors = F)
```

```r
cat('--- before trim ---\n')
dim(adt_raw)
dim(gex_raw)
```

```r
## trim raw data 

# make feature_id unique -- it's weird that they even have protein_id overlaps 
adt_raw = adt_raw %>% distinct(gene, .keep_all=TRUE) %>% remove_rownames() %>% column_to_rownames("gene")
gex_raw = gex_raw %>% distinct(gene, .keep_all=TRUE) %>% remove_rownames() %>% column_to_rownames("gene")

# remove isotype adt, HTOs in ADT
rm_idx = which(str_detect(rownames(adt_raw), regex("Isotype|HTO|isotype")) == TRUE)
adt_raw = adt_raw[-rm_idx, ]

# simplify protein_id
simplified_id = unlist(regmatches(rownames(adt_raw), regexpr("anti-|anti-human_", rownames(adt_raw)), invert = TRUE))
simplified_id = simplified_id[which(nchar(simplified_id) != 0)]
rownames(adt_raw) = simplified_id

cat('-- after trim --\n')
dim(adt_raw)
dim(gex_raw)
```

```r
## transform to Seurat obj

seurat_obj = CreateSeuratObject(counts = gex_raw, min.cells=0, min.features=0)
seurat_obj[['CITE']] = CreateAssayObject(counts = adt_raw)
cat('-- before QC --\n')
seurat_obj

# QC according to the paper 
seurat_obj[["percent.mt"]] = PercentageFeatureSet(seurat_obj, pattern = '^MT-')
VlnPlot(seurat_obj, features = c("nFeature_RNA", "nCount_RNA", "percent.mt"), ncol = 3)
seurat_obj <- subset(seurat_obj, subset = nFeature_RNA < 200 & nCount_RNA < 1000 & percent.mt > 10, invert = TRUE)
cat('-- After QC --\n')
seurat_obj

## it seems that their raw matrix already has QC done 
```

```r
# log1p normalize to gex & protein_expr as in the paper

DefaultAssay(seurat_obj) = 'CITE'
seurat_obj = NormalizeData(seurat_obj, normalization.method = "LogNormalize", scale.factor = 1e3)

DefaultAssay(seurat_obj) = 'RNA'
seurat_obj = NormalizeData(seurat_obj, normalization.method = "LogNormalize")
```

```r
# prepare protein_id, celltype label, array name, into right formats

cell.type = factor(str_split(colnames(seurat_obj), pattern='\\.', simplify = TRUE)[,4])
seurat_obj$cell.type = cell.type

# check celltype
table(cell.type)

# check features
DefaultAssay(seurat_obj) = 'CITE'
rownames(seurat_obj)
```

```r
# output to h5ad file 
# slim down seurat obj
DefaultAssay(seurat_obj) = 'RNA'
seurat_rna = DietSeurat(
  seurat_obj,
  counts = TRUE, # so, raw counts save to adata.layers['counts']
  data = TRUE,   # so, log1p counts save to adata.X when scale.data = False, else adata.layers['data']
  scale.data = FALSE, # if only scaled highly variable gene, the export to h5ad would fail. set to false
  #features = VariableFeatures(seurat_tmp_select), # export all genes, not just top highly variable genes
  assays = "RNA",
  #dimreducs = c("adt.umap"),
  graphs = NULL, # to RNA_nn -> distances, RNA_snn -> connectivities
  misc = TRUE
)

DefaultAssay(seurat_obj) = 'CITE'
seurat_adt = DietSeurat(
  seurat_obj,
  counts = TRUE, # so, raw counts save to adata.layers['counts']
  data = TRUE,   # so, log1p counts save to adata.X when scale.data = False, else adata.layers['data']
  scale.data = FALSE, # if only scaled highly variable gene, the export to h5ad would fail. set to false
  assays = "CITE",
  #dimreducs = c("adt.umap"),
  graphs = NULL,
  misc = TRUE
)

cat('---- RNA Assay ----\n')
seurat_rna

cat('\n---- CITE Assay ----\n')
seurat_adt
```

```r
SaveH5Seurat(seurat_rna, filename = "./data/sorted_cells.gex.raw.h5Seurat", overwrite=TRUE)
Convert("./data/sorted_cells.gex.raw.h5Seurat", dest = "h5ad", overwrite = TRUE)
```

```r
SaveH5Seurat(seurat_adt, filename = "./data/sorted_cells.adt.raw.h5Seurat", overwrite=TRUE)
Convert("./data/sorted_cells.adt.raw.h5Seurat", dest = "h5ad", overwrite = TRUE)
```
