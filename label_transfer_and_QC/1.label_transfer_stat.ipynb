# label transfer stat using seurat labeltransfer tools
## label transfer between Brescia acute covid dataset and covid-flu dataset
## stat using predicted scores for each cell cluster

```r
library(Seurat)
library(matrixStats)
library(plyr)
library(tidyverse)
```

```r
# prediction results
predictions = readRDS("labeltransfer_predictions_wcovidflu_celltype.rds")
```

```r
perct_to_predict <- data.frame(table(predictions$predicted_id, predictions$covid_flu_celltype)) %>%
  left_join(data.frame(table(predictions$predicted_id)), by = c("Var1")) %>%
  mutate(prct = Freq.x/Freq.y*100) %>%
  group_by(Var1) %>%
  filter(prct == max(prct))
```

```r
head(perct_to_predict)
```

```r
dim(perct_to_predict)
```

```r
mean_max <- predictions %>%
    group_by(predicted_id_matched) %>%
    summarise(mean_max = mean(prediction.score.max, na.rm = TRUE))
```

```r
perct_to_predict <- perct_to_predict %>% left_join(mean_max, by = c("Var1" = "predicted_id_matched"))
```

```r
head(perct_to_predict)
```

```r
cor.test(perct_to_predict$mean_max, perct_to_predict$prct, method = "pearson")
```

```r
ggplot(perct_to_predict, aes(x = mean_max, y = prct)) +
    geom_point(color = "blue", size = 3)+
    geom_smooth(method = "lm", se = TRUE) + 
    geom_text(aes(label = Var1))
```

```r
options(repr.plot.width=12, repr.plot.height=12)
ggplot(predictions, aes(x=prediction.score.max, fill=predicted_id_matched))+
    geom_density()+
    facet_wrap(~predicted_id_matched)
```

```r
# a good assigned population, seems the score reflective of the truth or not
CD8_Naive = filter(predictions, predicted_id_matched == "CD8_Naive")
CD8_Naive$correctly_assigned = (CD8_Naive$predicted_id_matched == CD8_Naive$covid_flu_celltype)
```

```r
ggplot(CD8_Naive, aes(x=prediction.score.max, fill=correctly_assigned))+
    geom_density()
```

```r
# a poorly assigned population, seems the score reflective of the truth or not
CD8_EM = filter(predictions, predicted_id_matched == "CD8_EM")
CD8_EM$correctly_assigned = (CD8_EM$predicted_id_matched == CD8_EM$covid_flu_celltype)
```

```r
ggplot(CD8_EM, aes(x=prediction.score.max, fill=correctly_assigned))+
    geom_density()
```
